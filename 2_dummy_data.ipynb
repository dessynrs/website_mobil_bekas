```python
# import library yg akan digunakan
from faker import Faker
from tabulate import tabulate

import random
from datetime import datetime, timedelta

import csv

import pandas as pd
```

```python
# define Faker yang digunakan (loc = ID)
FAKER = Faker('id_ID')
```

```python
# membuat tabel
def show_data(table):
    """
    Fungsi untuk menampilkan data

    arg:
        - table (dict) : data dictionary yang ingin ditampilkan

    return:
        None
    """

    tab = tabulate(tabular_data = table,
                   headers = table.keys(),
                   tablefmt = "psql",
                   numalign = "center")
    print(tab)
```

```python
def save_to_csv(data, nama_file):
    '''
    Fungsi untuk menyimpan data dummy ke csv

    args:
        - data (list)     : list of dictionary data yang akan dijadikan csv
        - nama_file (str) : nama untuk file csv
	
	return:
		- None
    '''
   
    # Membuat file csv
    with open(file = f"{nama_file}.csv", mode = 'w', newline = '') as csv_file:
        # Membuat writer csv
        writer = csv.writer(csv_file)

        # write header csv
        writer.writerow(list(data.keys()))
        
        # mengetahui panjang data
        len_data = len(list(data.items())[0][1])
        
        # write data ke file csv
        for i in range(len_data):
            row = []
            for key in data.keys():
                row.append(data[key][i])
            writer.writerow(row)
```

## users (sellers & buyers)

```python
# generate kolom nama
def generate_name(n_name):
    """
    Fungsi untuk membuat nama dummy

    arg:
        - n_name (int) : jumlah data nama yang ingin dibuat

    return:
        names (list) : list nama yang sudah dibuat
    """

    names = list()

    while len(names) < n_name:

        first_name = FAKER.first_name()
        last_name = FAKER.last_name()

        full_name = (f'{first_name} {last_name}')
        if full_name not in names:
            names.append(full_name)

    return names
```

```python
# tabel users
def users(n_cust, is_print):
    """
    Fungsi untuk membuat dummy data customer table
    header:
        - user_id
        - first_name
        - last_name
        - phone                
        - city_id
        - email

    arg:
        - n_cust (int)  : Jumlah customer yang ingin dibuat
        - is_print  (bool) : Jika True akan menampilkan hasil data

    return:
        - table  (list) :  
    """
    
    # Buat table
    table = {}
    table["user_id"] = [i+1 for i in range(n_cust)]
    names = generate_name(n_cust)
    table['first_name'] = [i.split(' ')[0] for i in names] 
    table['last_name'] = [i.split(' ')[1] for i in names] 
    table['phone'] = [FAKER.phone_number() for i in range(n_cust)]
    table['city_id'] = [random.choice(city['kota_id']) for i in range(n_cust)]
    table['email'] = [f"{name.lower().replace(' ', '')}@{FAKER.free_email_domain()}" \
                      for name in names]


    # Print table
    if is_print:
        show_data(table)

    return table
```

```python
# membuat data table seller
sellers = users(n_cust = 40,
                is_print = True)
```

```python
# menyimpan data customer dalam bentuk csv
# sellers
save_to_csv(data = sellers,
            nama_file = 'sellers')
```

```python
# untuk generate date
from random import randrange
from datetime import timedelta

def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
def product_table(n_product, is_print):
    """
    Fungsi untuk membuat dummy data product table : product

    header:
        - product_id
        - seller_id
        - brand
        - model
        - car_type
        - body_type
        - year
        - price
        - kilometer
        - date_post


    arg:
        - n_product (int)  : Jumlah customer yang ingin dibuat
        - is_print  (bool) : Jika True akan menampilkan hasil data

    return:
        - table  (list) :  
    """
    
    # Buat table
    table = {}
    table["product_id"] = car_product_detail['product_id']
    table['seller_id'] = [random.choice(sellers['user_id']) for i in range(n_product)]
    table['brand'] = car_product_detail['brand']
    table['model'] = car_product_detail['model']
    # car_type = ["Manual", "Matic"]
    table['car_type_id'] = [random.randint(1,2) for i in range(n_product)]
    table['body_type_id'] = car_product_detail['body_type_id']
    table['year'] = car_product_detail['year']
    table['price'] = car_product_detail['price']
    table['kilometer'] = [random.randint(75000,150000) for i in range(n_product)]
    
    d1 = datetime.strptime('1/1/2023 1:30 PM', '%m/%d/%Y %I:%M %p')
    d2 = datetime.strptime('10/31/2023 4:50 AM', '%m/%d/%Y %I:%M %p')
    table['date_post'] = [random_date(d1, d2) for i in range(n_product)]


    # Print table
    if is_print:
        show_data(table)

    return table
```

```python
# membuat data table product
n_product = car_product_detail['product_id'].count()
product_table = product_table(n_product = car_product_detail['product_id'].count(),
                              is_print = True)
```

```python
# menyimpan data customer dalam bentuk csv
save_to_csv(data = product_table,
            nama_file = 'product_table')
```

## bid_logs

```python
# bid_logs
def bid_logs(n_bid, is_print):
    """
    Fungsi untuk membuat dummy data bid_log

    header:
        - bid_id
        - buyer_id
        - product_id


    arg:
        - n_bid (int)  : Jumlah data bid yang ingin dibuat
        - is_print  (bool) : Jika True akan menampilkan hasil data

    return:
        - table  (list) :  
    """
    
    # Buat table
    table = {}
    table["bid_id"] = [i+1 for i in range(n_bid)]
    table["buyer_id"] = [random.choice(buyers['user_id']) for i in range(n_bid)]
    table["product_id"] = [random.choice(product_table['product_id']) for i in range(n_bid)]

    # Print table
    if is_print:
        show_data(table)

    return table
```

```python
# membuat data table bid_logs
bid_logs = bid_log(n_bid = 200,
                  is_print = True)
```

```python
bid_logs = pd.DataFrame(bid_logs)
bid_logs.head()
```

```python
# melengkapi data bid_logs

bid_merge = pd.merge(left=bid_logs, right=product_table, how='left', right_on='product_id', left_on='product_id')

bid_merge['date_post_dt'] = pd.to_datetime(bid_merge['date_post'])
bid_merge['date_bid'] = bid_merge['date_post_dt'] + timedelta(days=random.randint(1,10))
bid_merge['pct'] = [random.randint(1,10) for i in range(bid_merge['bid_id'].count())]
bid_merge['bid_price'] = bid_merge['price'] - (bid_merge['pct'] * bid_merge['price'] / 100)
bid_merge

bid_logs_detail = bid_merge[['bid_id', 'buyer_id', 'product_id', 'date_bid', 'bid_price']]
bid_logs_detail['bid_price'] = bid_logs_detail['bid_price'].astype('int')
bid_logs_detail.head()
```

```python
# menyimpan data customer dalam bentuk csv
save_to_csv(data = bid_logs_detail,
            nama_file = 'bid_logs')

# mengecek isi file yang sudah dibuat
bid_logs = pd.read_csv('bid_logs.csv', delimiter = ',')
bid_logs.head()
```
